- **RMSE** (Root Mean Squared Error): 예측값과 실제값의 평균 제곱근 오차 (낮을수록 좋음)
- **MAE** (Mean Absolute Error): 예측값과 실제값의 평균 절대 오차 (낮을수록 좋음)
- **R²** (R-squared): 결정 계수, 모델의 설명력 (1에 가까울수록 좋음, 0~1)
